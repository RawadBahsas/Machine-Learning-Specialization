```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1,2])
y_train = np.array([300,500])
print(f'the feature array is {x_train}')
print(f'the target array is {y_train}')
```

```python
m = x_train.shape[0]
print(f'the training set size is {m}')
```

```python
%matplotlib inline
# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r')
plt.show()
```

```python
def compute_model_output(x,w,b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w*x[i]+b
        
    return f_wb
```

```python
w=200
b=100
tmpFwb = compute_model_output(x_train,w,b)
plt.plot(x_train,tmpFwb,c='b')
plt.scatter(x_train,y_train,marker='x',c='r')
plt.show()
```
