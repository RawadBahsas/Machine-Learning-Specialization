```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load our data set
x_train = np.array([1.0, 2.0,3.0,5.6,9.8])   #features
y_train = np.array([300.0, 500.0,600.0,750.0,1000.0])   #target value
```

```python
def squared_error_cost_function(x, y, w, b):    
    cost = 0
    m = x.shape[0]
    for i in range(m):
        cost += ((w*x[i]) + b - y[i])**2
        
    return cost/2*m
        
```

```python
jCost = squared_error_cost_function(x_train,y_train,200,100)
print(jCost)
```

```python
def calc_derivative_gradient_w(x,y,w,b):    
    m = x.shape[0]
    der = 0
    for i in range(m):
        der+= ((w*x[i])+b-y[i])*x[i]
    return der/m

def calc_derivative_gradient_b(x,y,w,b):    
    m = x.shape[0]
    der = 0
    for i in range(m):
        der+= ((w*x[i])+b-y[i])
    return der/m
```

```python
gradient = calc_derivative_gradient_b(x_train,y_train,200,100)
print(gradient)
```

```python
def gradient_descent_function(x, y, itirations, alpha, in_w, in_b):
    w = in_w
    b = in_b   
    j_cost_array = []
    points_array=[]
    for i in range (itirations):
            tmp_w = w - alpha * calc_derivative_gradient_w(x,y,w,b)
            tmp_b = b - alpha * calc_derivative_gradient_b(x,y,w,b)
            w=tmp_w
            b=tmp_b
            cost = squared_error_cost_function(x,y,w,b)
            j_cost_array.append(cost)
            points_array.append((w,b))
    return w,b,j_cost_array,points_array
```

```python
alpha = 1.0e-2
w,b,jcost,pointsarr = gradient_descent_function(x_train,y_train,10000,alpha,0,0)
print(w.round())
print(b.round())
```
